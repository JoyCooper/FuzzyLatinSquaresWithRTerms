```python
from sage.combinat.matrices.latin import *
from itertools import combinations
import itertools
import numpy as np
np.set_printoptions(threshold=np.inf)
%run ./combinations_of_permutations.ipynb
```

```python
square_size = 5
max_dimension = factorial(square_size) - (square_size-1)^2
L_start = back_circulant(square_size)
permutation_list = genPermutationsBetween(square_size,square_size)
latin_vector_list = [combination_of_permutations.latin_square(L_start).make_vector(permutation_list)]
Latin_matrix = Matrix(QQ,latin_vector_list[0])
for latin in random_latin_generator:
    CoP_latin = combination_of_permutations.latin_square(latin)
    vector_latin = CoP_latin.make_vector(permutation_list)  
    M2 = Latin_matrix.stack(vector_latin)
    if M2.rank() > Latin_matrix.rank():
        Latin_matrix = M2
        latin_vector_list.append(vector_latin)
        print(Latin_matrix.rank())
        if Latin_matrix.rank() >= max_dimension:
            break
    else:
        soln = Latin_matrix.solve_left(vector_latin)
        coeff_list = []
        term_list = []
        for i in range(len(soln)):
            if soln[i] != 0:
                coeff_list.append(soln[i])
                term_list.append(combination_of_permutations.vector(latin_vector_list[i],permutation_list))
        print(coeff_list)
        for term in term_list:
            print(term)
```

```python
for latin in latin_vector_list:
    cop = combination_of_permutations.vector(latin,permutation_list)
    print(f"{cop.nested_list()},")
```
